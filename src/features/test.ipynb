```python
img_list = glob.glob("/Volumes/ExFAT/road_shoulder_gan/data/raw/gsv/image/perspective/COH_97mXOJbZnxsHDUfJTA*.png")
img_agg = None
for i in range(0,360,90):
    # get img that match with the direction in each loop
    img_file_match = list(filter(lambda x: f"Direction_{str(i)}" in x, img_list))[0]
    img_temp = cv2.cvtColor(cv2.imread(img_file_match), cv2.COLOR_BGR2RGB)
    if img_agg is None:
        img_agg = img_temp
    else:
        img_agg = cv2.hconcat([img_agg, img_temp])
plt.imshow(img_agg)
plt.xticks([]), plt.yticks([])  # Hides the graph ticks and x / y axis
plt.show()  
```
